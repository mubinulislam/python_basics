```python
x=int(input("enter number 1: "))
y=int(input("enter number 2: "))
try:
  z=x/y
except Exception as e:
  print("Exception occured: ",e)
  z=None
print("Division: ",z)
```

```python
x=input("enter number 1: ")
y=input("enter number 2: ")
try:
  z=int(x)/y
except Exception as e:
  print("Exception occured: ",type(e).__name__) #if dont know the exception name then use <type(e).__name__> to know the name.
  z=None
print("Division: ",z)
```
